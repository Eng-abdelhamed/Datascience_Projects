```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel(r'/content/Employees.xlsx')
```

```python
df
```

```python
df.drop('No',axis=1,inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
df['Gender'].value_counts(normalize=True) * 100
```

```python
plt.figure(figsize=(14,8),edgecolor='red')
# plt.fon
plt.subplot(1,2,1)
sns.barplot(x = df['Gender'].value_counts().index,
            y=df['Gender'].value_counts(),color ='mediumseagreen')
plt.title("Gender Distrribution")
plt.xlabel('Gender')
plt.legend()
plt.tight_layout()
plt.ylabel('Frequecy')
gen = df['Gender'].value_counts()
for i in range(len(gen)):
  plt.text(i,gen[i] + 2,gen[i] , ha ='center' , fontsize = 'large')

plt.subplot(1,2,2)
plt.pie(x= df['Gender'].value_counts() ,
        labels = df['Gender'].value_counts().index ,
        autopct = '%1.2f%%',
        explode =[0,0.2]
        ,shadow = True,radius =1
        )
plt.title("GENDER Distribution")
plt.legend()
plt.tight_layout()
plt.show()
```
